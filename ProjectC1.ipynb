```python
interest_counts = data['On a scale of 1 to 5, with 1 being not at all interested and 5 being extremely interested, how interested are you in taking more computing classes?'].value_counts().sort_index()
```

```python
#Questions 
#1. On a scale of 1 to how many students have an interest in taking computer classes?
#2. Gender among students taking comp sci classes.
#3. Ages of the students who are taking the computer classes.
#4. Race of students taking computer science classes.
```

```python
#Plots for the questions
```

```python
plt.figure(figsize=(8, 4))
interest_counts.plot(kind='bar', color='red')
plt.title('Interest in Taking More Computing Classes')
plt.xlabel('Interest Level')
plt.ylabel('Number of Students')
plt.show()
```

```python
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(8, 4))
gender_counts.plot(kind='pie', colors=['lightblue', 'lightgreen'])
plt.title('Distribution of Gender Among Students')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
data['Age '].hist(bins=20, color='orange')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
race_counts = data['Race/ethnicity'].value_counts()
plt.figure(figsize=(10, 6))
race_counts.plot(kind='pie', colors=['lightblue', 'yellow'])
plt.title('Distribution of Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Frequency')
plt.show()
```
