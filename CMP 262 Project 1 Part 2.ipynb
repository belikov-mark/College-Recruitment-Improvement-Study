```python
#Changing the column name
```

```python
computing_df.columns = computing_df.columns.str.lower().str.replace(' ', '_')
```

```python
non_computing_df.columns = non_computing_df.columns.str.lower().str.replace(' ', '_')
```

```python
#getting rid of redundance
```

```python
print(computing_df.columns)
```

```python
ifeatures = ['gender', 'race/ethnicity', 'age_']
```

```python
computing_df.drop(columns=ifeatures, inplace=True)
```

```python
non_computing_df.drop(columns=ifeatures, inplace=True)
```

```python
#condeses the majors
```

```python
def Condense(Major):
    if 'computer science' in Major.lower():
        return 'Computer Science'
    elif 'information technology' in Major.lower():
        return 'Information Technology'
    elif 'software engineering' in Major.lower():
        return 'Software Engineering'
    else:
        return 'Other'
```

```python
#adding the major to the column
```

```python
computing_df['major'] = computing_df['what_degree_program_are_you_currently_enrolled_in?'].apply(Condense)
```

```python
non_computing_df['major'] = non_computing_df['what_degree_program_are_you_currently_enrolled_in?'].apply(Condense)
```
